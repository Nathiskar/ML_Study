Label Encoding :


*   converting the labels into numeric form




```python
# importing the Dependencies
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
# loading the data from csv file to pandas data frame
cancer_data = pd.read_csv('/content/breast_cancer_data.csv')
```

```python
# first 5 rows of the dataFrame
cancer_data.head()
```

```python
# finding the count of the different labels
cancer_data['diagnosis'].value_counts()
```

```python
# load the LabelEncoder Function
label_encode = LabelEncoder()
```

```python
label = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
# appending the labels to the dataFrame
cancer_data['target'] = label
```

```python
cancer_data.head()
```

0 --> Bengin
1 --> Malignant

```python
cancer_data['target'].value_counts()
```

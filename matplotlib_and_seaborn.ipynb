#Matplotlib

  -> Useful for making Plots

```python
# importing matplotlib library
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plot
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show
```

```python
# adding title x - axis and y - axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine wave')
plt.title('sine wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y, 'b*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Parabola')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'c^')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
ax.bar(languages, people)
plt.xlabel('languages')
plt.ylabel('people')
plt.show()
```

Pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 150, 40, 70]
ax.pie(people, labels = languages, autopct = '%1.1f')
plt.show()
```

Scatter plot


```python
x = np.linspace(0,10,50)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

3D Scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, c=z, cmap='Blues')
```

# Seaborn
  Data Visualization Library

Importing the library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note:  Seaborn has some built in dataset

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
#visualize the data
sns.relplot(data=tips, x = 'total_bill', y = 'tip', col = 'time',hue = 'smoker', style='smoker', size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter Plot

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data=iris)
```

```python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

Count plot

```python
titanic.shape
```

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(x='survived', data=titanic)
```

Bar chart

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
# house price dataset
from sklearn.datasets import load_wine
wine = load_wine() # not be in pandas dataset
print(wine)
wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
wine['price'] = wine.target
```

```python
wine_df.head()
```

Distribution Plot

```python
sns.displot(wine_df['proline'])
```

```python
sns.distplot(wine_df['proline'])
```

Correlation

  1. positive correlation
  2. negative correlation

Heat Map

```python
correlation = wine_df.corr()
```

```python
# constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
plt.show()
```

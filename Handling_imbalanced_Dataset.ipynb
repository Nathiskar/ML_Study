Imbalanced Dataset

A dataset with an unequal class distribution

```python
# importing the dependencies
import numpy as np
import pandas as pd
```

```python
# loading the dataset to the data frame
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
# first 5 row of the dataframe
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# distribution of the 2 classes
credit_card_data['Class'].value_counts()
```

This is highly imbalanced dataset



0 -> legal


1 -> non legal

```python
# separating the legit and fradual transcation
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class ==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-sampling

Building a sample dataset containing similar distribution of legit and Fradulent Transcation

Number of Fradual Transcation -> 492

```python
legit_sample = legit.sample(n=103)
```

```python
print(legit_sample.shape)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
new_dataset.head()
```
